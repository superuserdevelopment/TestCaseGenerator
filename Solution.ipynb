Harry and Sejal are extremely happy to have their first child. Their son is in love with toys, so Harry wants to buy some. In front of him lies a number of different toys, tagged with their prices. Harry has only a certain amount to spend, and he wants to maximize the number of toys he's buying with that money.

Given a list of prices and an amount to spend, what is the maximum number of toys Harry can buy? 

For example, if  prices = **[1,2,3,4]** (prices: an array of integers representing toy prices) and Harry has  **k = 7** (k: an integer, Harry's budget) to spend, he can buy items **[1,2,3]** for 6, or  **[3,4]** for 7 units of currency. He would choose the first group of 3 items.

**Input Format**<br>
The first line contains **one integer, n**, the number of priced toys.<br>
The next line contains **one integer, k**, the maximum amount Harry has to spend<br>
The next line contains **n space-separated integers** prices[i].<br>

**Output Format**<br>
An integer that denotes the maximum number of toys Harry can buy for his son.

```python
def maximumToys(prices, k):
    prices.sort()
    s = 0
    l = 0
    for i in range(len(prices)):
        if s > k:
            break
        s = s + prices[i]
        l = l + 1
    return(l - 1)
maximumToys([1, 12, 5, 111, 200, 1000, 10], 50)
```

```python
def minimumBribes(Q):
    
    #
    # initialize the number of moves
    moves = 0 
    #
    # decrease Q by 1 to make index-matching more intuitive
    # so that our values go from 0 to N-1, just like our
    # indices.  (Not necessary but makes it easier to
    # understand.)
    Q = [P-1 for P in Q]
    #
    # Loop through each person (P) in the queue (Q)
    for i,P in enumerate(Q):
        
        if P - i > 2:
            return("Too chaotic")
        
        for j in range(max(P-1,0),i):
            if Q[j] > P:
                moves += 1
    return(moves)
minimumBribes([2, 5, 1, 3, 4])
```

```python
import collections as co
def riddle(arr):
    stack = co.deque()
    max_window = {}
    arr += [0,]
    for i, n in enumerate(arr):
        if not stack or n > stack[-1][1]:
            stack.append((i, n))
        else:
            while stack and stack[-1][1] >= n:
                if i-stack[-1][0] not in max_window or max_window[i-stack[-1][0]] < stack[-1][1]:
                    max_window[i-stack[-1][0]] = stack[-1][1]
                p = stack.pop()
            stack.append((p[0], n))
    ans = [0]*(len(arr)-1)
    v_ant = min(max_window.values())
    for i in range(len(arr)-1, 0, -1):
        if i in max_window and max_window[i] > v_ant:
            ans[i-1] = max_window[i]
            v_ant = max_window[i]
        else:
           ans[i-1] = v_ant
    return ans
riddle([46, 76, 52, 77, 90, 48, 88, 28, 43, 69, 23, 7, 58, 20, 90, 71, 45, 77, 29, 27, 21, 70, 27, 21, 84, 83, 6, 70, 41, 65, 11, 45, 45, 15, 85, 88, 1, 78, 80, 85, 51, 91, 64, 53, 64, 2, 68, 0])
```
